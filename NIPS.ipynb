```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('NIPS_1987-2015.csv')
df
```

```python
df.shape
```

```python
df1 = df.iloc[:, 1:]
df1
```

```python
# the scipy version to produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

X = linkage(df1,'ward')

plt.figure()
dendrogram(X)
plt.show()
```

```python
df_t = df.T
df_t
```

```python
labels = df_t.iloc[0]
df_t.columns = labels
df_t
```

```python
df3_t = df_t.iloc[1:,:]
df3_t
```

```python
Y = linkage(df3_t,'ward')

plt.figure(figsize=(20,15))
dendrogram(Y)
plt.show()
```
